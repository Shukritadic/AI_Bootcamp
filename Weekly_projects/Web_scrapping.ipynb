```python
import requests
from bs4 import BeautifulSoup

# Step 1: Define your target website and the data you want to scrape
url = "https://www.jumia.co.ke/health-beauty/"
# Define the data you want to scrape (e.g., computer_names)
data_to_scrape = []

# Step 2: Send HTTP requests to the target website's server to retrieve the HTML or XML content
response = requests.get(url)

# Step 3: Parse the HTML response
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')

    # Step 4: Extract the desired data
    # Find all the prices of computers
    computer_prices = soup.find_all('div', class_='name')
#computer_prices
    # Extract and store the titles
    for name in computer_prices:
        # Extract text content of the computer_names and strip any leading/trailing whitespace
        names= name.text.strip()

        # Append the extracted title to the list of data to scrape
        data_to_scrape.append(computer_prices)
#else:
   # print('Failed to fetch webpage:', response.status_code)

# Print or further process the scraped data
#print("scrapped computer prices :")
for name in data_to_scrape:
    print(names)
```
